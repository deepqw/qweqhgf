```python
device = (
    "cuda"
    if torch.cuda.is_available()
    else "cpu"
)
print(f"Using {device} device")
```

```python
transform = v2.Compose([
    v2.Resize((256, 256)),
    # v2.CenterCrop(224),
    # v2.ToTensor(),
    v2.Grayscale(),
    v2.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
    #v2.Normalize(mean=[0.5, 0.5, 0.5], std=[0.5, 0.5, 0.5])
])
```

```python
class Dataset(torch.utils.data.Dataset):
    def __init__(self, paths, labels=None, transform=None, dir_path = ''):
        super().__init__()
        self.paths = paths
        self.labels = labels
        self.transform = transform
        self.dir_path = dir_path

    def __len__(self):
        return len(self.paths)

    def __getitem__(self, ind):

        # img = torchvision.io.read_image(self.dir_path + self.paths[ind]) / 255

        img = Image.open(self.dir_path + self.paths[ind]).convert('RGB') #.convert('L')
        # img = np.array(img.convert('RGB')).reshape(3, 256, 256) # RGB image
        # img = np.array(img).reshape(1, 256, 256) # grayscale image
        # img = torch.FloatTensor(img / 255)

        if self.transform:
            img = self.transform(img)

        # label = self.labels[ind]

        if self.labels is not None:
            return img, self.labels[ind]
        else:
            return img
```

```python
dataset = Dataset(
    paths = img_train_df['id'].tolist(),
    dir_path = '/content/Train/',
    transform = models.ResNet18_Weights.IMAGENET1K_V1.transforms(),
    labels = img_train_df['label'].tolist()
)
```

```python
dataloader = DataLoader(dataset, batch_size=64)
```

```python
next(iter(dataloader))[1].shape
```

```python
next(iter(dataloader))[0].shape
```

```python
model = models.resnet18(weights='IMAGENET1K_V1')
model.fc = nn.Linear(512, 8, bias=True)
```

```python
optimizer = torch.optim.Adam(model.parameters(), lr=1e-3)
# optimizer = torch.optim.Adam(model.parameters(), lr=3e-4)
loss_fn = torch.nn.CrossEntropyLoss()
```

```python
model.train()
model.to(device)
epochs = 10

for epoch in range(epochs):  # loop over the dataset multiple times

    running_loss = 0.0
    i=0
    for data in tqdm(dataloader):
        i+=1
        # get the inputs; data is a list of [inputs, labels]
        # print(data[0])
        # print(data[1])
        inputs, labels = data[0].to(device), data[1].to(device)

        # zero the parameter gradients
        optimizer.zero_grad()

        # forward + backward + optimize
        outputs = model(inputs)
        loss = loss_fn(outputs, labels)
        loss.backward()
        optimizer.step()

        # print statistics
        running_loss += loss.item()
        if i % 20 == 19:    # print every 20 mini-batches
            print(f'[{epoch + 1}, {i + 1:5d}] loss: {running_loss / 20:.3f}')
            running_loss = 0.0

print('Finished Training')
```

```python
testdataset = Dataset(
    paths = test_df['id'].tolist(),
    dir_path = '/content/Test/',
    transform = models.ResNet18_Weights.IMAGENET1K_V1.transforms()
)
```

```python
def predict(image):
    model.eval()
    X = image
    with torch.no_grad():
        X = X.to(device)
        y = model(X).argmax(-1).cpu().item()
    return y
```

```python
pred = [predict(image.reshape(1, 3, 224, 224)) for image in tqdm(testdataset)]
pred[:5]
```
