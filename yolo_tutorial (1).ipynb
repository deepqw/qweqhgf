```python
import pandas as pd
import numpy as np
from ultralytics import YOLO
import rarfile
import zipfile as zf
import torch
from PIL import Image
```

```python
files = zf.ZipFile('archive.zip')
files.extractall()
```

```python
rar = rarfile.RarFile('data.rar')
rar.extractall()
```

```python
if torch.cuda.is_available(): device = 'cuda'
else: device = 'cpu'
print(f"Using {device} device")
```

```python
model = YOLO('yolov8n.pt')
```

```python
train_results = model.train(data='data.yaml', epochs=10, imgsz=256, device=device, plots=True)
```

```python
val_results = model.val(device=device, plots=True)
```
