```python
import pandas as pd
import numpy as np

from surprise import Dataset
from surprise import Reader
from surprise.model_selection import train_test_split

from surprise import SVD, KNNBasic
from surprise.model_selection import GridSearchCV
from surprise.model_selection import cross_validate

from surprise import accuracy
```

```python
data = pd.read_csv('/content/drive/MyDrive/всерос_по_ии_24/тренировочный_этап/[M] Рекомендация курсов для студентов/train.csv')
data
```

```python
data['rating'] = 1
data
```

```python
# Определение формата данных для Surprise
reader = Reader(rating_scale=(0, 1))  # Предполагается, что course_id неявно означает положительный рейтинг (можно изменить при необходимости)

# Загрузка данных в Surprise Dataset
dataset = Dataset.load_from_df(data[['user_id', 'course_id', 'rating']], reader)

# Разделение данных на обучающую и тестовую выборки
trainset, testset = train_test_split(dataset, test_size=0.2)
```

```python
# Создание модели SVD
svd = SVD(n_factors=1, n_epochs=15000, lr_all=0.005, reg_all=0.02)
knn = KNNBasic()

# svd.fit(trainset)
knn.fit(trainset)
# results = cross_validate(svd, dataset, measures=['RMSE', 'MAE'], cv=5, verbose=True)
```

```python
# Получение прогнозов модели
# predictions = svd.test(testset)
predictions = knn.test(testset)

# Расчет MAE и RMSE
mae = accuracy.mae(predictions)
rmse = accuracy.rmse(predictions)
```

```python
model = knn
# model = svd
```

```python
def get_recommendations(user_id, top_n=3):
    """Возвращает список из top_n course_id, которые пользователь еще не слушал."""

    # Получение списка всех курсов
    all_courses = set(data['course_id'].unique())

    # Получение списка курсов, которые пользователь уже слушал
    user_courses = set(data[data['user_id'] == user_id]['course_id'].values)

    # Получение списка курсов, которые пользователь еще не слушал
    unseen_courses = all_courses - user_courses
    # unseen_courses = all_courses

    # Предсказание рейтингов для всех unseen_courses
    predictions = [model.predict(user_id, course_id) for course_id in unseen_courses]

    # Сортировка предсказаний по убыванию рейтинга
    predictions.sort(key=lambda x: x.est, reverse=True)

    # Возвращение top_n course_id
    return [prediction.iid for prediction in predictions[:top_n]]

# Пример использования: получение рекомендаций для пользователя с user_id = 1
recommendations = get_recommendations(1)
recommendations
```

```python
sub = pd.read_csv('/content/drive/MyDrive/всерос_по_ии_24/тренировочный_этап/[M] Рекомендация курсов для студентов/sample_submission.csv')
sub
```

```python
for i, id in enumerate(sub.user_id):

    if id in data.user_id:
        sub['course_id_1'][i] = get_recommendations(id)[0]
        sub['course_id_2'][i] = get_recommendations(id)[1]
        sub['course_id_3'][i] = get_recommendations(id)[2]

    else:
        sub['course_id_1'][i] = 7
        sub['course_id_2'][i] = 1
        sub['course_id_3'][i] = 15

sub
```
