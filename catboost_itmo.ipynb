```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from catboost import CatBoostClassifier
```

```python
train = pd.read_csv("Downloads/train.csv")
train
```

```python
test = pd.read_csv("Downloads/test.csv")
test
```

```python
train.isna().sum().sum()
```

```python
train = train.dropna()
train.isna().sum().sum()
```

```python
train = train.drop(["index"], axis=1)
train
```

```python
#Ð¿Ñ€Ð¾Ð²ÐµÑ€ÐºÐ° Ð½Ð° Ð´ÑƒÐ±Ð»Ð¸ÐºÐ°Ñ‚Ñ‹
train.duplicated().sum()
```

```python
test.isna().sum()
```

```python
for i in test.columns[20:]:
  test[i] = test[i].fillna(test[i].mean())
```

```python
test.isna().sum().sum()
```

```python
X_test = test.drop(["index"], axis=1)
X_test
```

```python
X_train = train.drop(["target"], axis=1)
X_train
```

```python
le = LabelEncoder()
for i in X_train.columns[:19]:
  X_train[i] = le.fit_transform(X_train[i])
  X_test[i] = le.fit_transform(X_test[i])
```

```python
X_train.info()
```

```python
X_test.info()
```

```python
le_target = LabelEncoder()
y_train = train.target
y_train = le_target.fit_transform(y_train)
y_train
```

```python
model = CatBoostClassifier(iterations=1000, 
                           learning_rate=0.1, 
                           depth=12, 
                           loss_function='MultiClass', 
                           random_state=42,
                           task_type='GPU')
```

```python
model.fit(X_train, y_train, verbose=200)
```

```python
testY = model.predict(X_test)
testY
```

```python
testY = le_target.inverse_transform(testY)
testY
```

```python
prediction = test[["index"]].copy()
prediction["target"] = testY
prediction
```
